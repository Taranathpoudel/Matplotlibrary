# The subplot function takes three arguements i.e.
# subplot(1,2,3) here 1 is No of rows 2 is No of columns and 3 is index 

# Actually subplot is a type of plot within a same graph

```python
import matplotlib.pyplot as plt
import numpy as np

# assigning values
x=np.array([1,2,3])
y=np.array([1,4,9])
plt.subplot(2,1,1)
plt.grid()
plt.plot(x,y)

# for secoand chart
x=np.array([5,6,7])
y=np.array([25,36,49])
plt.subplot(2,1,2)
plt.grid()
plt.plot(x,y)

plt.show()
```
