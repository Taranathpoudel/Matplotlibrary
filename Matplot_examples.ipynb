# Rainfall Data

```python
import matplotlib as plt
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,lw=0.5,color='r')
plt.xlabel('Time in hrs')
plt.ylabel('Intensity of rainfall')
plt.grid()
```

# Particle Size Distribution of All in Aggregate


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

y=[100,100,91.9,62.72,33.081,28.671,19.137,5.772,2.336,1.322]
x=[106,50,20,11.2,4.75,2.36,1.18,0.6,0.3,0.15]
from scipy.ndimage.filters import gaussian_filter1d

ysmoothed = gaussian_filter1d(y, sigma=2)
plt.semilogx(x, ysmoothed)
plt.xlabel('Particle Size Distribution')
plt.ylabel('% Finer')
plt.grid()
plt.show()
```

# Working with Numpy

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,6])
y=np.array([3,8])
plt.plot(x,y)
plt.show()
```

# Plotting Values without a line
