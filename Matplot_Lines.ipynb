# All about lines

```python
# 'linewidth or lw keyword can be used for line width'
import matplotlib.pyplot as plt
import numpy as np

x=np.array([20,30,40,50,60])
y=np.array([400,900,1600,2500,3600])
plt.semilogx(x,y,lw='1')
plt.grid()
plt.show()
```

# Plotting Multiple lines

```python
# Taking Above Programme as reference plt.plot() goives one line
x1=[1,3,5,7,9]
x2=[2,4,6,8,10]
plt.plot(x1)
plt.plot(x2)
plt.grid()
plt.show()
```

# Plotting Separate Lines defined in array

```python
x1=np.array([1,10,20,30,40])
y1=np.array([2,3,5,25,4])
x2=np.array([90,100,80,85])
y2=np.array([35,40,45,50])

plt.plot(x1,y1,x2,y2,'o--')
plt.grid()
plt.show()
```
