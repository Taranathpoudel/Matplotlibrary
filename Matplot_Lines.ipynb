# All about lines

```python
# 'linewidth or lw keyword can be used for line width'
import matplotlib.pyplot as plt
import numpy as np

x=np.array([20,30,40,50,60])
y=np.array([400,900,1600,2500,3600])
plt.semilogx(x,y,lw='1')
plt.grid()
plt.show()
```

# Plotting Multiple lines

```python
# Taking Above Programme as reference plt.plot() goives one line
x1=[1,3,5,7,9]
x2=[2,4,6,8,10]
plt.plot(x1)
plt.plot(x2)
plt.grid()
plt.show()
```
